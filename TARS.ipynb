# DATA PREPROCESSING 

```python
#IMPORTING LIBRARIES
import pandas as pd


#IMPORTING THE DATASET
df= pd.read_excel(r'C:\Users\hp\Desktop\SEMESTER 3\MBD 511 MODELLING IN OPERATIONS MANAGEMENT\project1\personalized ad\TARS\Online Retail.xlsx')
print(df.head())
print( "COLUMN NAMES ARE:\n",df.columns)
```

```python
#EXTRACTING JUST THOSE COLUMNS THAT WE WILL NEED TO WORK ON AND DROPPING THE REMAINING COLUMNS. ALSO REMOVING EMPTY VALUES.
df=df[['CustomerID','InvoiceDate','Description']].dropna()
print(df.dtypes)
print(df.head())
```

# FEATURE EXTRACTION AND TABLE CREATION FOR EACH CUSTOMER

```python
#STORING ALL THE TABLES AS VALUES IN A DICTIONARY WHOSE KEYS ARE THE CORRESPONDING CUSTOMER ID 
dictionary_of_tables={}

grp=df.groupby('CustomerID')
#A GROUP OBJECT WILL HAVE TWO COMPONENTS. ONE:THE VALUE OF THE ATTRIBUTE ON WHICH THE DATASET IS GROUPED, TWO: THE CRRESPONDING GROUPS FOR EACH UNIQUE VALUE OF THAT ATTRIBUTE.
for id_, group in grp:          
    l=list()
    for date, group_ in group.groupby('InvoiceDate'):
        tuple_=date,group_['Description'].tolist()
        l.append(tuple_)
        
    dictionary_of_tables[id_]=pd.DataFrame(l,columns=['TIMESTAMP','BASKET'])
```
